```python
import pandas as pd

# 1. Load the data
# Use the path relative to your notebooks/ folder
df = pd.read_csv('../data/raw/online_retail.csv', encoding='ISO-8859-1')

# 2. Data Cleaning & Preparation
df.dropna(subset=['CustomerID'], inplace=True)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
# Convert period to string immediately for easier narrative generation later
df['YearMonth'] = df['InvoiceDate'].dt.to_period('M').astype(str)

# 3. Monthly Performance Summary (KPIs)
monthly_summary = df.groupby('YearMonth').agg({
    'TotalAmount': 'sum',
    'InvoiceNo': 'nunique',
    'CustomerID': 'nunique'
}).reset_index()

# Rename columns for clarity
monthly_summary.columns = ['YearMonth', 'TotalRevenue', 'TotalTransactions', 'UniqueCustomers']

# 4. Country-Based Performance (Top 10)
country_summary = df.groupby('Country')['TotalAmount'].sum().sort_values(ascending=False).head(10).reset_index()
country_summary.columns = ['Country', 'TotalRevenue']

# 5. Best Selling Products (Top 10)
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10).reset_index()
top_products.columns = ['ProductDescription', 'TotalQuantitySold']

# Check results
print("Monthly Summary Head:")
print(monthly_summary.head())
```

```python
knowledge_base = []

# Generate Monthly Insights
for _, row in monthly_summary.iterrows():
    knowledge_base.append(
        f"In {row['YearMonth']}, the business generated a total revenue of {row['TotalRevenue']:.2f} "
        f"across {row['TotalTransactions']} transactions from {row['UniqueCustomers']} unique customers."
    )

# Generate Country Insights
for _, row in country_summary.iterrows():
    knowledge_base.append(
        f"The country {row['Country']} is a top market with a total revenue of {row['TotalRevenue']:.2f}."
    )

# Save to processed folder
with open('../data/processed/knowledge_base.txt', 'w', encoding='utf-8') as f:
    for line in knowledge_base:
        f.write(line + "\n")
```

```python
import pandas as pd

# Load the cleaned data from your EDA process
# Assuming the file is in the data/raw or data/processed directory
df = pd.read_csv('../data/raw/online_retail.csv', encoding='ISO-8859-1')

# Basic cleaning for the narrative generation
df.dropna(subset=['CustomerID'], inplace=True)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
df['YearMonth'] = df['InvoiceDate'].dt.to_period('M')

# KPI 1: Monthly Performance (Revenue, Orders, Customers)
monthly_kpi = df.groupby('YearMonth').agg({
    'TotalAmount': 'sum',
    'InvoiceNo': 'nunique',
    'CustomerID': 'nunique'
}).reset_index()

# KPI 2: Top 10 Countries by Revenue
country_performance = df.groupby('Country')['TotalAmount'].sum().sort_values(ascending=False).head(10).reset_index()

# KPI 3: Top 10 Best Selling Products
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
import os

knowledge_base_records = []

# Convert Monthly KPIs to sentences
for _, row in monthly_kpi.iterrows():
    sentence = (f"During {row['YearMonth']}, the business generated a total revenue of {row['TotalAmount']:.2f} "
                f"from {row['InvoiceNo']} invoices and {row['CustomerID']} unique customers.")
    knowledge_base_records.append(sentence)

# Convert Country Performance to sentences
for _, row in country_performance.iterrows():
    sentence = f"The country {row['Country']} is a top performer with a total lifetime revenue of {row['TotalAmount']:.2f}."
    knowledge_base_records.append(sentence)

# Convert Product Trends to sentences
for _, row in top_products.iterrows():
    sentence = f"The product '{row['Description']}' is a best-seller with a total of {row['Quantity']} units sold."
    knowledge_base_records.append(sentence)

# Save the narrative data to the processed folder
output_path = '../data/processed/knowledge_base.txt'
with open(output_path, 'w', encoding='utf-8') as f:
    for record in knowledge_base_records:
        f.write(record + "\n")

print(f"Success! Knowledge base generated at: {output_path}")
```
