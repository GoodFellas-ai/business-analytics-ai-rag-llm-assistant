```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter

# 1. Load the knowledge base
loader = TextLoader("../data/processed/knowledge_base.txt")
documents = loader.load()

# 2. Split text
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=0, separator="\n")
docs = text_splitter.split_documents(documents)

# 3. Initialize Updated Embeddings
# Connection error devam ederse VPN veya farklı bir internet dene
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# 4. Create and persist
persist_directory = "../data/processed/chroma_db"
vector_db = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    persist_directory=persist_directory
)

print("Vector store created successfully!")
```
